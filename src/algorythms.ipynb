```
# importing necessary libs
from pandas import DataFrame
import matplotlib.pyplot as plt
from random import uniform
```

```
# creating randomized arrays
data = {
    'x': [],
    'y': []
     }

# Creates the function to randomize arrays for the dict named "data"
def rd_array(array):
    for array_data in range(1, 100):
        array.append(uniform(1, 100))

rd_array(data['x'])
rd_array(data['y'])

print(data['x'])
print(data['y'])
```

```
df = DataFrame(data, columns=['x', 'y'])
df.head()
```

```
# imports the lib for the algorythm
from sklearn.cluster import KMeans
```

```
km = KMeans(n_clusters=2)  # Creates an object for the kmeans algorythm to find
km.fit(df)  # Applies the algorythm.                                2 clusters.
centroids = km.cluster_centers_  # Stores the km centroids in a variable.
print(centroids)
```

```
# Plots our data
plt.scatter(data['x'], data['y'], s=50, c=km.labels_.astype(float), alpha=1.0);
plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='red');
plt.xlabel('x');
plt.ylabel('y');
```

```
# Imports other libs
from sklearn import neighbors, datasets
import pandas as pd
import numpy as np
```

```
# Creates the routine to use the iris dataset
iris = datasets.load_iris()
```

```
# Converts iris data into a datafre
df_iris = pd.DataFrame(np.c_[iris['data'], iris['target']],
                       columns=iris['feature_names'] + ['targets'])

# Display the iris dataframe head
df_iris.head()
```

```
# Converts the data into an array
x = df_iris.iloc[:, :-1].values
y = df_iris.iloc[:, 4].values
print(f'''X ARRAY (INPUT):
{x}
Y ARRAY (TARGET, OUTPUT):
{y}''')
```

```
# Divides the data into train and test

 # Function that divides the dataset
from sklearn.model_selection import train_test_split 
# Divides 20% of the data for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()  # Object that normalizes the data
scaler.fit(x_train)  # Normalizes the data

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

print(f'''
x_train:
{x_train}
x_test:
{x_test}
''')
```

```
# Training the model
from sklearn.neighbors import KNeighborsClassifier

# Builds the classification with 5 neighbors
classifier = KNeighborsClassifier(n_neighbors=5)  
classifier.fit(x_train, y_train)  # Applies the classification

print(classifier)
```

```
 # Prediction
y_pred = classifier.predict(x_test)
print(y_pred)
```

```
from sklearn.metrics import classification_report, confusion_matrix

# Builds the confunsion matrix to compare with the created model
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```
from mlxtend.plotting import plot_confusion_matrix

# Plots the confusion_matrix
y_confusion = confusion_matrix(y_test, y_pred)

fig, ax = plot_confusion_matrix(conf_mat=y_confusion)
fig.show()
```

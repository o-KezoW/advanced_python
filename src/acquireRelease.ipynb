# **PHILOSOFERS**

```
import threading
import time
import random
 
 
class Philosofers(threading.Thread):
    def __init__(self, name, left_fork, right_fork):
        print(f'{name} sat on the table')
        threading.Thread.__init__(self, name=name)
        self.left_fork = left_fork
        self.right_fork = right_fork
    
    def run(self):
        thread_name = threading.currentThread().getName()
        print(f'{thread_name} started to think')
        for times in range(1):
            time.sleep(random.randrange(5))
            print(f'{thread_name} stopped thinking')
            self.left_fork.acquire()
            time.sleep(1)
            try:
                print(f'{thread_name} got the left fork')
                self.right_fork.acquire()
                try:
                    print(f'{thread_name} got both forks and started eating')
                finally:
                    self.right_fork.release()
                    print(f'{thread_name} released the right fork')
            finally:
                self.left_fork.release()
                print(f'{thread_name} released the left fork')
 
 
fork_1, fork_2, fork_3, fork_4, fork_5, fork_6 = (
    threading.RLock(), threading.RLock(), threading.RLock(),
    threading.RLock(), threading.RLock(), threading.RLock(),
)
 
philo_1, philo_2, philo_3, philo_4, philo_5, philo_6 = (
    Philosofers('Aristotle', fork_1, fork_2),
    Philosofers('Karl Marx', fork_2, fork_3),
    Philosofers('Friedrich Nietzsche', fork_3, fork_4),
    Philosofers('Plato', fork_4, fork_5),
    Philosofers('Ludwig Wittgenstein', fork_5, fork_6),
    Philosofers('Socrates', fork_6, fork_1),
)
 
def thread_work(iterable, method):
    """
Iterates over an iterable and runs or stop threads
methods can be either "start" or stop
    """
    for thread in iterable:
        try:
            if method == 'start':
                thread.start()
            elif method == 'stop':
                thread.join()
        except:
            return f'"{thread}" is not a thread'
 
 
 
philo_list = (philo_1, philo_2, philo_3, philo_4, philo_5, philo_6)
 
thread_work(philo_list, method='start')
thread_work(philo_list, method='stop')
 
print('EVERYTHING IS DONE ALREADY')
```
